```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Cargar las imágenes
img_1 = cv.imread("DSC_1347.jpg")
img_2 = cv.imread("DSC_1347.png")

# Redimensionar las imágenes para que tengan el mismo tamaño
img_1_resized = cv.resize(img_1, (img_2.shape[1], img_2.shape[0]))

# Unir las dos imágenes horizontalmente
combined_image = np.concatenate((img_1_resized, img_2), axis=1)

# Mostrar las imágenes
# Mostrar la imagen combinada
plt.imshow(cv.cvtColor(combined_image, cv.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
